```python
import pandas as pd
df=pd.read_csv(r"C:\Users\USER\Downloads\movies_metadata.csv")
df.info()
```

```python
df.head()
```

```python
df= df[df["status"] == "Released"]
df.head()
```

```python
df=df.drop(columns=['belongs_to_collection','genres','homepage','imdb_id','original_title','spoken_languages','tagline','title','overview','poster_path','production_companies','production_countries','status','id'])
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["original_language"])
df["runtime"] = df["runtime"].fillna(df["runtime"].mean())
df=df.drop(columns=["release_date"])
df.head()
```

```python
df.info()
```

```python
df["budget"] = df["budget"].astype(int)
```

```python
df = df[df["budget"] != 0]
```

```python
df.info()
```

```python
df["popularity"] = df["popularity"].astype(float)
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

categorical_cols=["adult", "original_language","video"]

for col in categorical_cols:
    df[col]=le.fit_transform(df[col])

df.head()
```

```python
df['profit']=df['revenue']-df['budget']
```

```python
df.head()
```

```python
df["success"] = (df["profit"] > 0).astype(int)
```

```python
df.head()
```

```python
x=df.drop(columns=['success','budget','profit'])
y=df.success
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
model1=linear_model.LogisticRegression()
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn import tree 
from sklearn.tree import DecisionTreeClassifier
```

```python
model2=DecisionTreeClassifier()
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```
