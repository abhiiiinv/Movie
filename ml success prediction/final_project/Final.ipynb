```python
import pandas as pd
df=pd.read_csv("movies_metadata.csv")
df.info()
```

```python
df.head()
```

```python
df= df[df["status"] == "Released"]
df.head()
```

```python
df=df.drop(columns=['belongs_to_collection','genres','homepage','imdb_id','original_title','spoken_languages','tagline','title','overview','poster_path','production_companies','production_countries','status','id'])
df.head()
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["original_language"])
df["runtime"] = df["runtime"].fillna(df["runtime"].mean())
df=df.drop(columns=["release_date"])
df.head()
```

```python
df.info()
```

```python
df["budget"] = df["budget"].astype(int)
```

```python
df = df[df["budget"] != 0]
```

```python
df.info()
```

```python
df["popularity"] = df["popularity"].astype(float)
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

categorical_cols=["adult",'original_language',"video"]

for col in categorical_cols:
    df[col]=le.fit_transform(df[col])

df.head()
```

```python
df['profit']=df['revenue']-df['budget']
```

```python
df.head()
```

```python
df["success"] = (df["profit"] > 0).astype(int)
```

```python
df.head()
```

```python
x=df.drop(columns=['success','revenue','profit'])
y=df.success
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=41)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
model1=linear_model.LogisticRegression()
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn import tree 
from sklearn.tree import DecisionTreeClassifier
```

```python
model2=DecisionTreeClassifier()
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
df["success"].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier(n_estimators=20)
model3.fit(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

```python
import pickle

name='trained_movie_model2.sav'
pickle.dump(model3,open(name,'wb'))
```

```python
feature_columns = x_train.columns  # or X_test.columns

importances = model2.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': feature_columns, 'Importance': importances})
print(feature_importance_df.sort_values(by="Importance", ascending=False))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted')  # Use 'weighted' for multiclass
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')

    print(f"Accuracy: {accuracy*100}")
    print(f"Precision: {precision*100}")
    print(f"Recall: {recall*100}")
    print(f"F1-Score: {f1*100}")

    return accuracy, precision, recall, f1
```

```python
# Example usage with a trained model
y_pred1 = model1.predict(x_test)
y_pred2 = model2.predict(x_test)
y_pred3 = model3.predict(x_test)

print("LogisticRegression Metrics:")
evaluate_model(y_test, y_pred1)

print("\n DecisionTree Metrics:")
evaluate_model(y_test, y_pred2)

print("\n RandomForest Metrics:")
evaluate_model(y_test, y_pred3)
```
